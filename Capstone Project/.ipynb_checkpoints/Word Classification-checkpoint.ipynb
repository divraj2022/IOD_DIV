```python
Train_data[['genres', 'genres_1', 'genres_2']] = Train_data['genres'].str.split(',', expand=True)
Train_data.head()
```

```python
Train_data_1 = Train_data.drop(columns=['tconst'
                         ,'titleType'
                         ,'originalTitle'
                         ,'isAdult'
                         , 'startYear'
                         ,'endYear'
                         ,'runtimeMinutes'
                         ,'genres_1'
                         ,'genres_2'])
                         
```

```python
Train_data_1
```

```python
corpus = Train_data_1['primaryTitle'].values
corpus
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names_out())
```

```python
Vector_Text = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
Vector_Text['OriginalText'] = pd.Series(corpus)
Vector_Text
```

```python
text_data = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
text_data['genres'] = Train_data_1['genres']
text_data.head()
```

```python
Target= text_data.columns[-1]
Predictors= text_data.columns[:-1]
X= text_data[Predictors].values
y= text_data[Target].values
```

```python
import sklearn.metrics as metrics
clf = LogisticRegression(C=5, solver='newton-cg',penalty='l2')
LOG=clf.fit(X,y)
pred=LOG.predict(X)
Test_Data=pd.DataFrame(data=X, columns=Predictors)
Test_Data['TargetVariable']=y
Test_Data['Prediction']=pred
print(metrics.classification_report(y, pred))
print(metrics.confusion_matrix(pred, y))
F1_Score=metrics.f1_score(y, pred, average='weighted')
print('Accuracy of the model on Testing Sample Data:', round(F1_Score,2))
```

```python
def genres_test(input_text):
    test = vectorizer.transform(input_text)
    Prediction=clf.predict(test)
    Result=pd.DataFrame(data=input_text, columns=['title'])
    Result['Prediction']=Prediction
    return(Result)
```

```python
movie_name=["Flores y perlas"]
predicted_genre = genres_test(input_text=movie_name)
predicted_genre
```
