### Manuplating the Data

```python

X = car_5.drop(['Price'], axis = 1)
y = car_5['Price']
```

```python
X
```

#### Splitting the data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

#### Printing the size of train and test data

```python
print("The size of the input train data is: {}".format(X_train.shape))
print("The size of the output train data is: {}".format(y_train.shape))
print("The size of the input test data is: {}".format(X_test.shape))
print("The size of the output test data is: {}".format(y_test.shape))
```

#### Encoding the data

```python
encoder = TargetEncoder(cols = 'Year')
```

```python
encoder.fit(X_train['Year'], y_train.to_frame()['Price'])
```

```python
X_train['Year'] = encoder.transform(X_train['Year'])
```

```python
X_test['Year'] = encoder.transform(X_test['Year'])
```

```python
X_train.head()
```

```python
from category_encoders import TargetEncoder, OneHotEncoder
encoder = TargetEncoder(cols = 'Brand')
encoder.fit(X_train['Brand'], y_train.to_frame()['Price'])
X_train['Brand'] = encoder.transform(X_train['Brand'])
X_test['Brand'] = encoder.transform(X_test['Brand'])
```

```python
X_train.head()
```

```python
encoder = TargetEncoder(cols = 'Model')
encoder.fit(X_train['Model'], y_train.to_frame()['Price'])
X_train['Model'] = encoder.transform(X_train['Model'])
X_test['Model'] = encoder.transform(X_test['Model'])
```

```python
X_train.head()
```

#### One Hotcoding

```python
encoder = OneHotEncoder()
encoder.fit(X_train[['Type','Gearbox','Fuel','Status','Color']])
one_hot_encoded_output_train = encoder.transform(X_train[['Type','Gearbox','Fuel','Status','Color']])
one_hot_encoded_output_test = encoder.transform(X_test[['Type','Gearbox','Fuel','Status','Color']])
```

```python
X_train = pd.concat([X_train, one_hot_encoded_output_train], axis = 1)
X_test = pd.concat([X_test, one_hot_encoded_output_test], axis = 1)
```

```python
X_train.drop(['Type','Gearbox','Fuel','Status','Color'], axis = 1, inplace = True)
X_test.drop(['Type','Gearbox','Fuel','Status','Color'], axis = 1, inplace = True)
```

```python
X_train.info()
```

#### Standardising the data

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_new = scaler.transform(X_train)
X_test_new = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import sklearn.metrics as metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```
